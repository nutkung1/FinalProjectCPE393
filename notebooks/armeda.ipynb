```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

data= pd.read_csv('../data/Electric_Vehicle_Population_Data.csv')
data=data.dropna()
```

```python

plt.figure(figsize=(12, 6))
top_city = data['City'].value_counts().head(10)

sns.barplot(x=top_city.index, y=top_city.values, palette='Set2')
plt.title('Top 10 Cities with Registered Electric Vehicles')
plt.xlabel('City')
plt.ylabel('Number of Registered EVs')
plt.xticks(rotation=45, ha='right')

for i, value in enumerate(top_city.values):
    plt.text(i, value + 300, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(12, 6))
top_makes = data['Make'].value_counts().head(10)

sns.barplot(x=top_makes.index, y=top_makes.values, palette='Set2')
plt.title('Top 10 Most Common Makes')
plt.xlabel('Make')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45, ha='right')

for i, value in enumerate(top_makes.values):
    plt.text(i, value + 300, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()


```
