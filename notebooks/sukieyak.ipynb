```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/Electric_Vehicle_Population_Data.csv')
```

```python
data.columns
```

```python
data['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].unique()
```

```python

sns.scatterplot(data=data, x='Base MSRP', y='Electric Range')
plt.title("Electric Range vs Base MSRP")
plt.xlabel("Base MSRP ($)")
plt.ylabel("Electric Range (Miles)")
#sns.scatterplot(data=data, x='Base MSRP', y='Electric Range', hue='Make')
plt.show()
```

```python
data[data['Base MSRP'] > 180000]
```

```python
sns.scatterplot(data=data[data['Electric Vehicle Type'] == 'Plug-in Hybrid Electric Vehicle (PHEV)' ], x='Base MSRP', y='Electric Range')
plt.title("Electric Range vs Base MSR(PHEV)")
plt.xlabel("Base MSRP ($)")
plt.ylabel("Electric Range (Miles)")
plt.show()
sns.scatterplot(data=data[data['Electric Vehicle Type'] == 'Battery Electric Vehicle (BEV)' ], x='Base MSRP', y='Electric Range')
plt.title("Electric Range vs Base MSRP(EV)")
plt.xlabel("Base MSRP ($)")
plt.ylabel("Electric Range (Miles)")
plt.show()
```
